### Logistic Regression Model to Predict a Successful Vaginal Birth After C-Section

```python
from pickle import load

x_train_unfiltered = load(
    open("../../../data/processed/vbac/X_train_prepared.pkl", "rb")
)
x_test_unfiltered = load(open("../../../data/processed/vbac/X_test_prepared.pkl", "rb"))
y_test = load(open("../../../data/processed/vbac/y_test.pkl", "rb"))
y_train = load(open("../../../data/processed/vbac/y_train.pkl", "rb"))

feature_importances_df = load(
    open("../../../data/processed/vbac/feature_importances_df.pkl", "rb")
)
```

We'll use our feature importances dataframe to select features with an importance > 0.01.

```python
x_train = x_train_unfiltered[
    :,
    feature_importances_df.loc[
        feature_importances_df["importance"] > 0.01
    ].index.tolist(),
]
x_test = x_test_unfiltered[
    :,
    feature_importances_df.loc[
        feature_importances_df["importance"] > 0.01
    ].index.tolist(),
]
```

With our data prepared, we'll now train a basic logistic regression model to predict successful VBACs.

```python
from sklearn.linear_model import LogisticRegression

log_reg_model = LogisticRegression(random_state=14, max_iter=1000)

log_reg_model.fit(x_train, y_train)
```

```python
predictions = log_reg_model.predict(x_test)
```

Our confusion matrix indicates that our model has a high false-negative rate.

```python
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, predictions)
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm, display_labels=log_reg_model.classes_
)
disp.plot()
plt.show()
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, predictions)

f1
```
