### Logistic Regression Model to Predict a Successful Vaginal Birth After C-Section

```python
from pickle import load

x_train = load(open("../../../data/processed/vbac/filtered_x_train_prepared.pkl", "rb"))
x_test = load(open("../../../data/processed/vbac/filtered_x_test_prepared.pkl", "rb"))
y_test = load(open("../../../data/processed/vbac/y_test.pkl", "rb"))
y_train = load(open("../../../data/processed/vbac/y_train.pkl", "rb"))
```

With our data prepared, we'll now train a basic logistic regression model to predict successful VBACs.

```python
from sklearn.linear_model import LogisticRegression

log_reg_model = LogisticRegression(random_state=14, max_iter=1000)

log_reg_model.fit(x_train, y_train)
```

```python
predictions = log_reg_model.predict(x_test)
```

Our model achieved an r-squared score of 26.7% accuracy.

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, predictions)

r2
```

Our confusion matrix indicates that our model has a high false-negative rate.

```python
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, predictions)
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm, display_labels=log_reg_model.classes_
)
disp.plot()
plt.show()
```

```python
from pickle import dump

dump(log_reg_model, open("./models/log_reg_model.pkl", "wb"))
```

Our logistic regression model is able to predict VBAC outcomes with low accuracy. A more complex model may have greater success at accurate predictions.
